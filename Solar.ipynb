```
import pandas as pd
```

```
##Reshaping data
df_belgium = pd.read_csv('/work/Data/Solar_Solar_atlis/GSA_Report_Belgium.csv')
df_belgium
df_belgium.rename(columns={'Unnamed: 0': 'Hour'}, inplace=True)
df_belgium
# Convert a single column to string
df_belgium['Hour'] = df_belgium['Hour'].astype(str)

# Assuming df_belgium is the DataFrame we are working with and it has a 'Hour' column and month data
# Step 1: Extract the starting hour from the range (e.g., '0 - 1' becomes '0')
df_belgium['Hour'] = df_belgium['Hour'].str.split('-').str[0].str.strip()

# Step 2: Convert 'Hour' to integer
df_belgium['Hour'] = df_belgium['Hour'].astype(int)
df_belgium.to_parquet('/work/Data/Solar_Solar_atlis/Belgiumflotingsolar.parquet')
```

```
Belgiumflotingsolar = pd.read_parquet('/work/Data/Solar_Solar_atlis/Belgiumflotingsolar.parquet')
Belgiumflotingsolar
Belgiumflotingsolar_df = Belgiumflotingsolar.melt(id_vars='Hour', value_vars=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

Belgiumflotingsolar_df.rename(columns={'variable': 'Month'}, inplace=True)

Belgiumflotingsolar_df['Year'] = 2022
Belgiumflotingsolar_df['timestamp'] = pd.to_datetime(Belgiumflotingsolar_df['Year'].astype(str) + '-' + Belgiumflotingsolar_df['Month'] + '-01') + pd.to_timedelta(Belgiumflotingsolar_df['Hour'], unit='h')
Belgiumflotingsolar_df.drop(columns=['Year', 'Hour'], inplace=True)
Belgiumflotingsolar_df.set_index('timestamp', inplace=True)

Belgiumflotingsolar_df.to_parquet('/work/Data/Solar_Solar_atlis/Belgiumflotingsolar_df.parquet')
Belgiumflotingsolar_df
```

```
Total_energy_BE = Belgiumflotingsolar_df['value'].sum()
Total_energy_BE
```

```
import seaborn as sns
import matplotlib.pyplot as plt


Belgiumflotingsolar_df['hour']=Belgiumflotingsolar_df.index.hour
Belgiumflotingsolar_df['month']=Belgiumflotingsolar_df.index.month

heatmap_data = Belgiumflotingsolar_df.pivot_table(index='hour', columns='month', values='value')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu")
#Customize the plot
plt.title('Heatmap of Total photovoltaic power output (kWh) Belgium', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Hour of the Day', fontsize=12)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

plt.show()
```
